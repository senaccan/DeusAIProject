##  LINEAR REGRESSION - LİNEER REGRESYON

```python
y=df["price"].values
x=df.drop("price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()

x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping

model=Sequential()

model.add(Dense(128,activation ="relu"))  
model.add(Dense(128,activation ="relu"))
model.add(Dense(128,activation ="relu"))
model.add(Dense(128,activation ="relu"))
model.add(Dense(128,activation ="relu"))
model.add(Dense(128,activation ="relu"))
model.add(Dense(128,activation ="relu"))

model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_train,y_train),batch_size=250,epochs=250)
```

```python
lostData=pd.DataFrame(model.history.history)
lostData.head(6)
```

```python
lostData.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

## THE PROCESS OF MAKING A PREDICTION - TAHMİN YAPMA İŞLEMİ

```python
predictionSequence=model.predict(x_test) 
```

```python
mean_absolute_error(y_test,predictionSequence) 
```

```python
plt.scatter(y_test,predictionSequence)
plt.plot(y_test,y_test,"g")
```
